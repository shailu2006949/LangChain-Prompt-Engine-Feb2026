```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
```

# **Task 1: Replace Hardcoded Prompts**

```python
# Task 1: Converting hardcoded prompt into PromptTemplate

template_text = "Explain {topic} in simple terms for beginners"

explain_template = PromptTemplate(
    input_variables=["topic"],
    template=template_text
)

# Test
print(explain_template.format(topic="Artificial Intelligence"))
print(explain_template.format(topic="Python Programming"))
```

# **Task 2: Multi-Input Prompt System**

```python
# Task 2: Prompt with multiple inputs

multi_input_template = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

# Test cases
print(multi_input_template.format(topic="AI", audience="beginners", tone="friendly"))
print(multi_input_template.format(topic="Python", audience="kids", tone="fun"))
print(multi_input_template.format(topic="Deep Learning", audience="engineers", tone="technical"))
```

# **Task 3: Prompt Variations Engine**

```python
# Task 3: Creating 3 different prompt styles

teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 interview questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

# **Task 4: ChatPromptTemplate System**

```python
# Task 4: Chat-based prompt system with roles

chat_template = ChatPromptTemplate.from_messages([
    ("system", "You are a {role}"),
    ("user", "Explain {topic}")
])

print(chat_template.format(role="teacher", topic="Neural Networks"))
print(chat_template.format(role="interviewer", topic="Neural Networks"))
print(chat_template.format(role="motivator", topic="Neural Networks"))
```

# **Task 5: Input Validation Layer**

```python
# Task 5: Validate audience and tone inputs

def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

# Test
print(validate_inputs("beginner", "fun"))
print(validate_inputs("kids", "friendly"))  # invalid → defaults
```

# **Task 6: Prompt Generator App**

```python
# Task 6: Generate prompts based on style

style_templates = {
    "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
    "interview": "Ask interview questions on {topic} for {audience} in a {tone} tone",
    "storytelling": "Explain {topic} as a story for {audience} in a {tone} way",
}

def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    template = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=style_templates[style]
    )

    return template.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```

# **Task 7: Template Reusability Test**

```python
# Task 7: One template, multiple inputs

reusable_template = PromptTemplate(
    input_variables=["topic", "audience"],
    template="Explain {topic} for {audience}"
)

inputs = [
    {"topic": "AI", "audience": "students"},
    {"topic": "Python", "audience": "developers"},
    {"topic": "Cloud", "audience": "engineers"},
    {"topic": "Data Science", "audience": "analysts"},
    {"topic": "Cyber Security", "audience": "beginners"},
]

for i in inputs:
    print(reusable_template.format(**i))
```
